```python
from lexicalrichness import LexicalRichness
from collections import Counter
import pandas as pd
```

```python
text_files = ["corpus/passus-i-sggk.txt", "corpus/passus-ii-sggk.txt", "corpus/passus-iii-sggk.txt", "corpus/passus-iv-sggk.txt"]
```

```python
data = []
```

```python
for filepath in text_files:
    with open(filepath, "r", encoding="utf-8") as f:
        print(f"Processing: {filepath}")
        text = f.read().lower()
        lex = LexicalRichness(text)
        
        data.append({
            "Filename": filepath,
            "Word Count": lex.words,
            "Unique Words": lex.terms,
            "TTR": lex.ttr,
            "RTTR": lex.rttr,
            "CTTR": lex.cttr,
            "MSTTR": lex.msttr(segment_window=25),
            "MATTR": lex.mattr(window_size=25),
            "MTLD": lex.mtld(threshold=0.72),
            "HD-D": lex.hdd(draws=42),
            "voc-D": lex.vocd(ntokens=50, within_sample=100, iterations=3),
            "Herdan's": lex.Herdan,
            "Summer's": lex.Summer,
            "Dugast's": lex.Dugast,
            "Maas'": lex.Maas,
            "Yule's K": lex.yulek,
            "Yule's I": lex.yulei,
            "Herdan's Vm": lex.herdanvm,
            "Simpson's D": lex.simpsond
        })

df = pd.DataFrame(data)
display(df)
```
