```python
import pandas as pd
import os
```

```python
file = 'word-count.csv'
df = pd.read_csv(file)
print(df.head())
```

```python
df['Rank'] = df.index + 1
print(df.head())
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(10, 7))
plt.loglog(df['Rank'], df['Count'], marker='.', markersize=1, linestyle='None')
plt.title("Zipf's Law Demonstration: Word Frequency vs. Rank (Log-Log Scale)")
plt.xlabel('Rank (log scale)')
plt.ylabel('Frequency (log scale)')
plt.legend()

# Add theoretical Zipf's Law line for comparison
# Theoretical freq = F_max / R
max_freq = df['Count'].max()
ranks = df['Rank']
theoretical_freq = max_freq / ranks
plt.loglog(ranks, theoretical_freq, color='red', 
           linestyle='--', label='Theoretical Zipf\'s Law')
plt.legend()

plt.tight_layout()
plt.show()
```
